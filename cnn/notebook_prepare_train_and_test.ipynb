# Create train and test folders

In the available dataset, we also have a boundary file that we can use to crop image.
Let's define the below function to extract x and y boundary.

```python
import xml.etree.ElementTree
def extract_boundary(df, index):
    e = xml.etree.ElementTree.parse(df.annotation_path[index]).getroot()
    object_ = e.findall("object")
    xmin = int(object_[0].find("bndbox").find("xmin").text)
    ymin = int(object_[0].find("bndbox").find("ymin").text)
    xmax = int(object_[0].find("bndbox").find("xmax").text)
    ymax = int(object_[0].find("bndbox").find("ymax").text)
    #box = (xmin, ymin, xmax, ymax)
    #box
    return xmin, ymin, xmax, ymax
```

## Custom selection

Let's select 5 dog breeds with in our dataset:
- n02088094-Afghan_hound
- n02094114-Norfolk_terrier 
- n02086646-Blenheim_spaniel
- n02093256-Staffordshire_bullterrier
- n02099712-Labrador_retriever

Our train/test images are splitted with 80% of train / 20% of test.

```python
import shutil

# Clean train and test folders first
shutil.rmtree(output_train_path, ignore_errors=True)
shutil.rmtree(output_test_path, ignore_errors=True)
os.mkdir(output_train_path)
os.mkdir(output_test_path)

# Define train size and test size with 80% of train and 20 % of test
train_size = 80
test_size = 20

# Dog breeds selection
images_labels = ['n02088094-Afghan_hound', 
                 'n02094114-Norfolk_terrier', 
                 'n02086646-Blenheim_spaniel',
                 'n02093256-Staffordshire_bullterrier',
                 'n02099712-Labrador_retriever']
n_classes = len(images_labels)

for classe in range(n_classes):
    label = images_labels[classe]
    # Create subset of image with label
    df_subset = df[df.label == label]
    df_subset.reset_index(inplace=True)
    os.mkdir(output_train_path + os.sep + label)
    os.mkdir(output_test_path + os.sep + label)
    # collect train image
    for i in range(train_size):
        image = Image.open(df_subset.image_path[i])
        # Crop image
        xmin, ymin, xmax, ymax = extract_boundary(df_subset, i)
        # Store image
        image = image.crop((xmin, ymin, xmax, ymax))
        image.save(output_train_path + os.sep + label + os.sep + str(i) + ".jpg")
    # collect test_size image
    for i in range(test_size):
        image = Image.open(df_subset.image_path[i + train_size])
        # Crop image
        xmin, ymin, xmax, ymax = extract_boundary(df_subset, i + train_size)
        # Store image
        image = image.crop((xmin, ymin, xmax, ymax))
        image.save(output_test_path + os.sep + label + os.sep + str(i) + ".jpg")
```
